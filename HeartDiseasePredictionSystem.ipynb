## Importing the important libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Importing the dataset

```python
heart_disease_dataset = pd.read_csv('/content/heart_disease_data.csv')
```

```python
# getting the first 10 rows of the dataset
heart_disease_dataset.head(10)
```

```python
# getting the shape of the dataset
heart_disease_dataset.shape
```

```python
# null verification
heart_disease_dataset.isnull().sum()
```

```python
# getting the stastical insights of the dataset
heart_disease_dataset.describe()
```

```python
# the mean of the target
heart_disease_dataset['target'].value_counts()
```

```python
# on groping the data on the basis of the target value and finding the mean of all the values
heart_disease_dataset.groupby('target').mean()
```

```python
# splitting the data set in the X and Y variable and eleminating the target column from the data
X = heart_disease_dataset.drop(columns='target',axis = 1)
Y = heart_disease_dataset['target']
```

```python
print(X,Y)
```

```python
# on standaridzing the values of the X variable i.e. dataset
scaler = StandardScaler()
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
X = standardized_data
Y = heart_disease_dataset['target']
```

```python
# splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
# The shape of the X, X_train, X_test
print(X.shape, X_train.shape, X_test.shape)
```

### Training the model in the Logistic Regression Model because it is a binary classification problem i.e. we get either 0 or 1 as the output


```python
regression_model = LogisticRegression()
```

```python
regression_model.fit(X_train,Y_train)
```

## Evaluating the model by its accuracy score in the training data

```python
# Accuracy Score for training data
X_train_prediction = regression_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
training_data_accuracy
```

```python
# Accuracy SCore for testing data
X_test_prediction = regression_model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
testing_data_accuracy
```

## Making a pridiction system out of the model trained

```python
# Define the correct feature names (must match training data)
feature_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']

# Input data
input_data = (67,1,0,125,254,1,1,163,0,0.2,1,2,3)

# Convert input data to a DataFrame
input_df = pd.DataFrame([input_data], columns=feature_names)

# Standardize input data
std_data = scaler.transform(input_df)

# Make prediction
prediction = regression_model.predict(std_data)

# Print result
if prediction[0] == 0:
    print("The person does not have heart disease")
else:
    print("The person has heart disease")
```
